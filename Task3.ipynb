# **Count Plot on each column**

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(8,10));
sns.countplot(data['Category'],ax=axs[0][0])
sns.countplot(data['Region'],ax=axs[0][1])
fig=sns.countplot(data['Ship Mode'],ax=axs[1][0])
sns.countplot(data['Segment'],ax=axs[1][1])
fig.set_xticklabels(fig.get_xticklabels(), rotation=50)
```

#**Distribution plot**

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(8,10))
sns.distplot(data['Profit'],ax=axs[0][0],color='green')
sns.distplot(data['Sales'],ax=axs[0][1],color='red')
sns.distplot(data['Quantity'],ax=axs[1][0],color='black')
sns.distplot(data['Discount'],ax=axs[1][1],color='brown')
axs[0][0].set_title('Profit',fontsize=18)
axs[0][1].set_title('Sales',fontsize=18)
axs[1][0].set_title('Quantity',fontsize=18)
axs[1][1].set_title('Discount',fontsize=18)
```

#**Category wise sales discount profit**

```python
data_category = data.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
data_category.head()
```

```python
data_category.plot.pie(subplots=True,autopct='%1.1f%%',labels=data_category.index,figsize=(20,20))
plt.title('Category wise Sales,profit,Discount')
```

Maximun sales and Profit obtain in **Technology.**

Minimun profit obtain in **Furniture**

#**Sub-Category wise Profit,Sales,Discount**

```python
data_subc = data.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
data_subc.head(10)
```

**Based on sales**

```python
plt.figure(figsize = (10,12))
plt.pie(data_subc['Sales'], labels = data_subc.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

Copier than machines have high sales.

**Based on Discount**

```python
plt.figure(figsize = (15,12))
plt.pie(data_subc['Discount'], labels = data_subc.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

**Binders , Machines and then tables have high Discount.**

#**Based on Profit**

```python
data_subc.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',
                                                              figsize= (10,5),
                                                              label=['Avg Sales Price($)','Profit($)'])
```

**COPIER : HIGHEST PROFIT AS WELL AS SELL**
